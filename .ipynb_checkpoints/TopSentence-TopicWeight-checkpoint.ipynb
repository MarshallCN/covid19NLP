## Example of visulize networkx

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_edges_from(
    [('A', 'B'), ('A', 'C'), ('D', 'B'), ('E', 'C'), ('E', 'F'),
     ('B', 'H'), ('B', 'G'), ('B', 'F'), ('C', 'G')])

val_map = {'A': 1.0,
           'D': 0.5714285714285714,
           'H': 0.0}

values = [val_map.get(node, 0.25) for node in G.nodes()]

# Specify the edges you want here
red_edges = [('A', 'C'), ('E', 'C')]
edge_colours = ['black' if not edge in red_edges else 'red'
                for edge in G.edges()]
black_edges = [edge for edge in G.edges() if edge not in red_edges]

# Need to create a layout when doing
# separate calls to draw nodes and edges
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, cmap=plt.get_cmap('jet'), 
                       node_color = values, node_size = 500)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, edgelist=red_edges, edge_color='r', arrows=True)
nx.draw_networkx_edges(G, pos, edgelist=black_edges, arrows=False)
plt.show()
```

# Extract mapped topic weights
从daily_news_balanced中，提取每一个mapped topic向量，并转化每一个topic为一列

```python
with open('./data_combined/daily_news_balanced.pickle', 'rb') as handle:
    daily_news_balanced = pickle.load(handle)
```

```python
d_news = daily_news_balanced[23].head(5)
d_news
```

```python
    sort_weighted = pd.DataFrame(0,columns=range(1,21),index=range(len(d_news)))
    dicts = d_news['weightedTopic']
    for i in tqdm(range(len(dicts))):
        if dicts[i] is not None:
            for k,v in dicts[i].items():
                sort_weighted.loc[i,k] = v
    sort_weighted
```

```python
for d in tqdm(range(23,30)):
    d_news = daily_news_balanced[d] 
    sort_weighted = pd.DataFrame(0, columns=range(1,21),index=range(len(d_news)))
    dicts = d_news['weightedTopic']
    for i in tqdm(range(len(dicts))):
        if dicts[i] is not None:
            for k,v in dicts[i].items():
                sort_weighted.loc[i,k] = v
    del(daily_news_balanced[d])
    with open(f'sort_weighted_{d}.pickle', 'wb') as handle:
        pickle.dump(sort_weighted, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
